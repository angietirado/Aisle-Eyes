```python
from pathlib import Path
from google.colab import files

uploaded = files.upload()
video_path = Path(list(uploaded.keys())[0])
print("Input:", video_path.resolve())
```

```python
from aisle_eyes.pipeline import run_pipeline

out_video = video_path.with_name(video_path.stem + "_aisle_eyes" + video_path.suffix)
out_csv = out_video.with_suffix(".csv")

# ROI as fractions of frame: left, top, right, bottom in [0, 1]
roi_frac = (0.32, 0.22, 0.68, 0.88)

run_pipeline(
    input_path=video_path,
    output_video=out_video,
    summary_csv=out_csv,
    model_name="yolov8n.pt",
    roi_pixels=None,
    roi_fractions=roi_frac,
)

print("Annotated video:", out_video)
print("CSV:", out_csv)
files.download(str(out_video))
files.download(str(out_csv))
```
